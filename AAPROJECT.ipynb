```python
import pandas as pd
import os

# Carregar o arquivo CSV
file_path = "C:\\Users\\JDC\\AA_Group14Project.csv"
df = pd.read_csv(file_path, delimiter=";")

# Garantir que as colunas de 'Electricity Produced (kWh)' e 'Global irradiance (W/m2)' sejam numéricas
df['Electricity Produced (kWh)'] = df['Produced Energy (kWh)'].str.replace(',', '.').astype(float)
df['Global irradiance (W/m2)'] = df['G(i)'].str.replace(',', '.').astype(float)

# Filtrar as linhas onde 'Global irradiance (W/m2)' >= 100 e 'Electricity Produced (kWh)' >= 2
df_filtered = df[(df['Global irradiance (W/m2)'] >= 100) & (df['Electricity Produced (kWh)'] >= 2)]

# Converter a coluna 'Date' para o formato datetime
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], format='%d/%m/%Y %H:%M', errors='coerce')

# Verificar se a conversão foi bem-sucedida (se houver valores nulos, corrija o formato)
df_filtered = df_filtered.dropna(subset=['Date'])  # Remover linhas com valores nulos em 'Date'

# Extrair apenas a data (sem hora) para o agrupamento diário
df_filtered['Date'] = df_filtered['Date'].dt.date

# Agrupar por 'Date' e calcular a média de 'Global irradiance (W/m2)' e 'Electricity Produced (kWh)' por dia
df_daily_avg = df_filtered.groupby('Date')[['Electricity Produced (kWh)', 'Global irradiance (W/m2)']].mean()

# Reorganizar a tabela para o layout desejado
df_daily_avg = df_daily_avg.reset_index()  # Reorganizar o índice

# Configurar para mostrar todas as linhas
pd.set_option('display.max_rows', None)

# Definir o caminho para salvar o novo arquivo CSV na mesma pasta que o arquivo original
output_path = os.path.join(os.path.dirname(file_path), 'Processed_Data.csv')

# Salvar o DataFrame filtrado e processado em um novo arquivo CSV
df_daily_avg.to_csv(output_path, index=False)

# Exibir todas as linhas da tabela
df_daily_avg.head(100)





```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Carregar o arquivo CSV processado
df = pd.read_csv("C:\\Users\\JDC\\Processed_Data.csv")

# Definir as variáveis independentes (G(i)) e dependentes (kWh)
X = df[['Global irradiance (W/m2)']]  # Variável independente
y = df['Electricity Produced (kWh)']  # Variável dependente

# Criar o modelo de regressão linear sem intercepto (passa pela origem)
model = LinearRegression(fit_intercept=False)

# Ajustar o modelo aos dados
model.fit(X, y)

# Coeficiente angular (declive)
slope = model.coef_[0]  # Declive

# R² (coeficiente de determinação) sem intercepto
r_squared = model.score(X, y)

# Exibir resultados com 5 casas decimais
print(f"Declive (coeficiente angular): {slope:.5f}")
print(f"R²: {r_squared:.5f}")

# Plotar gráfico de dispersão com a linha de regressão
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Dados')
plt.plot(X, model.predict(X), color='red', label=f'Regressão Linear: y = {slope:.5f}x')
plt.xlabel('Global irradiance (Wh/m²)')  # Alterado para "Wh/m²"
plt.ylabel('Electricity Produced (kWh)')
plt.title('Relação entre G(i) e kWh Produzidos')
plt.legend()
plt.show()



```

```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("C:\\Users\\JDC\\AA_Group14Project.csv", delimiter=";")

# Garantir que as colunas sejam numéricas
df['G(i)'] = df['G(i)'].str.replace(',', '.').astype(float)
df['Produced Energy (kWh)'] = df['Produced Energy (kWh)'].str.replace(',', '.').astype(float)

# Filtrar as linhas onde kWh é maior que 0
df_non_zero = df[df['Produced Energy (kWh)'] > 0]

# Multiplicar os valores de G(i) por 0.03533
df_non_zero['Adjusted kWh'] = df_non_zero['G(i)'] * 0.03533

# Somar os valores ajustados
total_estimated_kwh = df_non_zero['Adjusted kWh'].sum()

# Somar todos os valores originais de kWh
total_original_kwh = df['Produced Energy (kWh)'].sum()

# Calcular a percentagem do erro
error_percentage = abs(total_estimated_kwh - total_original_kwh) / total_original_kwh * 100

# Exibir os resultados
print(f"Total estimado de kWh produzidos: {total_estimated_kwh:.5f}")
print(f"Total original de kWh: {total_original_kwh:.5f}")
print(f"Erro percentual: {error_percentage:.2f}%")
```

```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("C:\\Users\\JDC\\TWOAXIS_SOLARTRACK.csv", delimiter=";")

# Substituir as vírgulas por pontos nas colunas que precisam de valores numéricos
df['G(i)'] = df['G(i)'].str.replace(',', '.').astype(float)
df['kWh'] = df['kWh'].str.replace(',', '.').astype(float)

# Multiplicar os valores de G(i) por 0.03533, exceto nas linhas onde o valor de kWh é 0
df['Calculated kWh'] = df.apply(lambda row: row['G(i)'] * 0.03533 if row['kWh'] != 0 else 0, axis=1)

# Somar todos os valores calculados de kWh
total_calculated_kWh = df['Calculated kWh'].sum()

# Somar todos os valores de kWh originais
total_original_kWh = df['kWh'].sum()

# Calcular o aumento percentual na produção
production_increase_percentage = (total_calculated_kWh - total_original_kWh) / total_original_kWh * 100

# Exibir os resultados
print(f"Total de kWh calculado: {total_calculated_kWh:.2f}")
print(f"Total de kWh original: {total_original_kWh:.2f}")
print(f"Aumento percentual na produção: {production_increase_percentage:.2f}%")
```
